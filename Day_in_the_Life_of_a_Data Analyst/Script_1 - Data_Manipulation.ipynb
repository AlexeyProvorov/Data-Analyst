```python
import pandas as pd
import os
```

```python
pwd = os.getcwd()
```

```python
dataset = pd.read_excel(
    pwd + "//Data - Survey Monkey Output.xlsx", sheet_name="Edited_data")
dataset.head()
```

```python
dataset_modified = dataset.copy()
```

```python
columns_to_drop = ['Start Date', 'End Date', 'Email Address',
                   'First Name', 'Last Name', 'Custom Data 1', ]
columns_to_drop
```

```python
dataset_modified = dataset_modified.drop(columns=columns_to_drop)
dataset_modified
```

```python
id_vars = list(dataset_modified.columns[0:8])
value_vars = list(dataset_modified.columns[8:])
```

```python
dataset_melted = dataset_modified.melt(
    id_vars=id_vars, value_vars=value_vars, var_name='Question+SubQuestion', value_name='Answer')
```

```python
dataset_melted.head()
```

```python
questions_dataset = pd.read_excel(
    pwd + "//Data - Survey Monkey Output.xlsx", sheet_name="Question")
```

```python
questions = questions_dataset.copy()
questions.shape
```

```python
list(questions.columns)
```

```python
questions.drop(columns = ['Raw Questions',
 'Raw Subquestions',
 'Subquestion'], inplace = True)
```

```python
questions
```

```python
dataset_merged = pd.merge(left=dataset_melted, right=questions, how='left',
         left_on="Question+SubQuestion", right_on="Question+Subquestion")
```

```python
respondent = dataset_merged[dataset_merged['Answer'].notna()]
respondent = respondent.groupby(["Question"])["Respondent ID"].nunique().reset_index()
respondent.rename(columns={"Respondent ID":"Respondents" }, inplace=True)
respondent.head()
```

```python
dataset_merged.shape
```

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=respondent, how='left',
                              left_on="Question", right_on="Question")
print("Orig Data", len(dataset_melted))
print("Merget Data", len(dataset_merged_two))
dataset_merged_two.head(3)
```

```python
same_answer = dataset_merged.groupby(
    ['Question+SubQuestion', 'Answer']).agg({'Respondent ID': 'count'}).reset_index()
same_answer.rename(columns={"Respondent ID": "Same Answer"}, inplace=True)
```

```python
dataset_merged_three = pd.merge(left=dataset_merged_two, right=same_answer, how='left', left_on=[
                                "Question+SubQuestion", "Answer"], right_on=["Question+SubQuestion", 'Answer'])
dataset_merged_three['Same Answer'].fillna(0, inplace=True)

print("Original Data", len(dataset_merged_two))
print("Merged Data", len(dataset_merged_three))
dataset_merged_three
```

```python
dataset_merged_three.columns
```

```python
output = dataset_merged_three.copy()
output.rename(columns={'Identify which division you work in.-Response':'Division_Primary',
       'Identify which division you work in.-Other (please specify)':'Division_Secondory',
       'Which of the following best describes your position level?-Response':"Position",
       'Which generation are you apart of?-Response':"Generation",
       'Please select the gender in which you identify.-Response':"Gender",
       'Which duration range best aligns with your tenure at your company?-Response':"Tenure",
       'Which of the following best describes your employment type?-Response':'Employment',
       },inplace=True)
output.head()
```

```python
output.to_excel(pwd + '//Final_output.xlsx', index=False)
```
